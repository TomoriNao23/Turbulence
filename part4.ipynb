#### Calculate 'Enstrophy'

```matlab
enst =calculateEnstrophy(X, Y, Z, u_turb, v_turb, w_turb);
enst = permute(enst,[1 3 2]);
xnew = X(1:end-1);
ynew = Y(1:end-1);
znew = Z(1:end-1);
[a,b,c] = meshgrid(xnew,znew,ynew);
```

```matlab
isosurface(a,b,c,enst,4)
print('./photo/part4/isosurface','-dpng')
```
