## <center>Part1:Solving __'Mean Volocity' 'Reynolds Stress' 'Moment'__

#### Import Data

```matlab
clc;clear;close all;
addpath('function','data');
load('import_phy.mat')
```

#### Calculate 'Mean Field' 'Disturbance Field' 'Reynolds Stress'

$\langle {u_l} \rangle (y_j)=\frac{1}{N_xN_z}\sum_{k=0}^{N_z-1}{\sum_{i=0}^{N_x-1}{u_l(x_i,y_j,z_k)}}$  
$\langle {u_l'u_m'} \rangle (y_j)=\frac{1}{N_xN_z}\sum_{k=0}^{N_z-1}{\sum_{i=0}^{N_x-1}{(u_l(x_i,y_j,z_k)-\langle u_l\rangle(y_j))(u_m(x_i,y_j,z_k)-\langle u_m \rangle(y_j))}}$

```matlab
[u_avg, v_avg, w_avg,p_avg] = calculateMeanVelocity(Up, Vp, Wp,P);
[u_turb, v_turb, w_turb,p_turb] = calculateTurbulentFluctuations(Up, Vp, Wp, P,u_avg, v_avg, w_avg,p_avg);
[rey_uu, rey_uv, rey_uw, rey_vv, rey_vw, rey_ww] = calculateReynoldsStress(u_turb, v_turb, w_turb);
```

#### Calculate 3rd 4th 'Moment'

$\langle {(u_l')^n} \rangle (y_j)=\frac{1}{N_xN_z}\sum_{k=0}^{N_z-1}{\sum_{i=0}^{N_x-1}{(u_l(x_i,y_j,z_k)-\langle u_l\rangle(y_j))^n}}$

```matlab
[u_3, v_3, w_3] = calculateNthMoments(u_turb, v_turb, w_turb, 3);
[u_4, v_4, w_4] = calculateNthMoments(u_turb, v_turb, w_turb, 4);
```

#### Plot 'Mean Volocity Profile'

```matlab
velocity = {u_avg; v_avg; w_avg};
colors = {'r', 'g', 'b'};
legend = {'$\langle{u}\rangle$', '$\langle{v}\rangle$', '$\langle{w}\rangle$'};
outputPath = './photo/part1/Mean Velocity Profile.png';
labelsx = 'y';
labelsy = '';
title = 'Mean Velocity Profile';

plot1(Y, velocity,colors,outputPath, labelsx, labelsy, title, legend)
```

#### Plot 'Reynolds Stress'

```matlab
rey_stresses = {rey_uu, rey_uv, rey_uw, rey_vv, rey_vw, rey_ww};
colors = {'g', 'y', 'm', 'b', 'r', 'c'};
legend = {
          '$\langle{u''u''}\rangle$',
          '$\langle{u''v''}\rangle$',
          '$\langle{u''w''}\rangle$',
          '$\langle{v''v''}\rangle$',
          '$\langle{v''w''}\rangle$',
          '$\langle{w''w''}\rangle$'
          }';
outputPath = './photo/part1/Reynolds_Stress.png';
labelsx = 'y';
labelsy = '';
title = 'Reynolds Stress';
plot1(Y, rey_stresses, colors, outputPath, labelsx, labelsy, title, legend);
```
